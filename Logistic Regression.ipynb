```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
```

```python
#crate a dataset 
x,y = make_classification(n_samples=100,n_features=10,n_classes=2,random_state=42)
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
#model traing
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
```

```python
logistic.fit(x_train,y_train)
```

```python
y_pred = logistic.predict(x_test)
print(y_pred)
```

```python
logistic.predict_proba(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = logistic.predict(x_test)
acc = accuracy_score(y_test, y_pred)
cnn = confusion_matrix(y_test, y_pred)
rep = classification_report(y_test, y_pred)

print("Accuracy:", acc)
print("Confusion Matrix:\n", cnn)
print("Classification Report:\n", rep)

print(rep)
```

## Hyperparameter Tuning and Cross Validation 

```python
model = LogisticRegression(max_iter=1000)
penalty = ['l1','l2','elasticnet']
c = [100,10,1.0,0.1,0.01]
solver = ['newton-cg','lbfgs','liblinear','sag','saga']
```

```python
params = dict(penalty = penalty,C= c,solver = solver)
params
```

```python
# Grid searchCV
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold(5)
grid = GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(x_test)

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = logistic.predict(x_test)
acc = accuracy_score(y_test, y_pred)
cnn = confusion_matrix(y_test, y_pred)
rep = classification_report(y_test, y_pred)

print("Accuracy:", acc)
print("Confusion Matrix:\n", cnn)
print("Classification Report:\n", rep)

print(rep)
```

## Randomized  searchCV

```python
import warnings
from sklearn.exceptions import ConvergenceWarning
warnings.filterwarnings("ignore", category=ConvergenceWarning)
from sklearn.model_selection import RandomizedSearchCV
```

```python
model  = LogisticRegression()

param_grid = {
    'penalty': ['l1', 'l2', 'elasticnet'],
    'C': [0.1, 1, 10],
    'solver': ['saga'],  # saga supports all penalties
    'l1_ratio': [0.5]     # needed only for elasticnet
}

randomcv = RandomizedSearchCV(estimator=model,param_distributions=param_grid,cv=5,scoring='accuracy')
```

```python
randomcv.fit(x_train,y_train)
```

```python
randomcv.best_score_
```

```python
randomcv.best_params_
```

```python
y_pred = randomcv.predict(x_test)
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = logistic.predict(x_test)
acc = accuracy_score(y_test, y_pred)
cnn = confusion_matrix(y_test, y_pred)
rep = classification_report(y_test, y_pred)

print("Accuracy:", acc)
print("Confusion Matrix:\n", cnn)
print("Classification Report:\n", rep)
print(rep)
```

## Logistic Regresssion For Multiclass Classification Problem

```python
x,y = make_classification(n_classes=3,n_features=10,n_informative=3,n_samples=1000,random_state=42)
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split

x_train ,x_test , y_train, y_test = train_test_split(x, y, test_size=0.03,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

logistic = LogisticRegression(multi_class='ovr')

logistic.fit(x_train,y_train)

y_pred = logistic.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = logistic.predict(x_test)
acc = accuracy_score(y_test, y_pred)
cnn = confusion_matrix(y_test, y_pred)
rep = classification_report(y_test, y_pred)

print("Accuracy:", acc)
print("Confusion Matrix:\n", cnn)
print("Classification Report:\n", rep)

print(rep)
```
