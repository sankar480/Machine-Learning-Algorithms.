**Linear Regreesion**

**Formula**

**y = mx+b**

y = Reffer to Dependent variable

m = Reffer to Slope value

x = Reffer to Independent variable

b = Reffer to Intercept value

```python
# import library
import numpy as np
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
# upload the dataset
canada = pd.read_csv('/content/canada_per_capita_income.csv')
canada.head()
```

```python
canada.isnull().sum()
```

```python
# plot data
plt.scatter(canada.year,canada.per_income,color='r')
plt.xlabel("Years")
plt.ylabel("Price")
plt.show()
```

```python
# Define features (independent variable) and target (dependent variable)
x = canada[['year']] # Features (Year must be in 2D for sklearn)
y = canada['per_income']
```

```python
# Train the linear regression model
reg = linear_model.LinearRegression()
reg.fit(x,y)
```

```python
# predict data
reg.predict([[2001]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
2001 * 828.46507522 + -1632210.7578554575
```

```python
# show the linear regression model line
plt.scatter(canada.year,canada.per_income,color='r')
plt.xlabel("Years")
plt.ylabel("Price")
plt.plot(canada.year,reg.predict(canada[['year']]),color = 'blue')
plt.show()
```
