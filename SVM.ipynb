```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
from sklearn.datasets import make_classification

x,y = make_classification(n_classes=2,n_features=2,n_samples=1000,n_redundant=0,n_clusters_per_class=1)
```

```python
x
```

```python
y
```

```python
x_df = pd.DataFrame(x)
```

```python
z = pd.DataFrame(x)[1]
```

```python
sns.scatterplot(x=x_df.iloc[:, 0], y=x_df.iloc[:, 1], hue=y)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x, y,test_size=0.03,random_state=42)
```

```python
from sklearn.svm import SVC

model = SVC(kernel='linear')

model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)

from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

acc = accuracy_score(y_pred,y_test)
con = confusion_matrix(y_pred,y_test)
report = classification_report(y_pred,y_test)

print(acc)
print(con)
print(report)
```

```python

x,y = make_classification(n_classes=2,n_features=2,n_samples=1000,n_redundant=0,n_clusters_per_class=2)
```

```python

x_train,x_test,y_train,y_test = train_test_split(x, y,test_size=0.03,random_state=42)
```

```python
rbf = SVC(kernel='rbf')
```

```python
rbf.fit(x_train,y_train)
```

```python
y_pred = rbf.predict(x_test)

from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

acc = accuracy_score(y_pred,y_test)
con = confusion_matrix(y_pred,y_test)
report = classification_report(y_pred,y_test)

print(acc)
print(con)
print(report)
```

```python
polynamial = SVC(kernel='sigmoid')
polynamial.fit(x_train,y_train)
y_pred1 = polynamial.predict(x_test)

from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

acc = accuracy_score(y_pred1,y_test)
con = confusion_matrix(y_pred1,y_test)
report = classification_report(y_pred1,y_test)

print(acc)
print(con)
print(report)
```
