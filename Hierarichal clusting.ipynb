```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import datasets
```

```python
iris = datasets.load_iris()
iris
```

```python
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_scaled = scaler.fit_transform(iris_df)
```

```python
x_scaled
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

pca_saled = pca.fit_transform(x_scaled)
```

```python
plt.scatter(pca_saled[:,0],pca_saled[:,1],c =iris.target)
```

```python
import scipy.cluster.hierarchy as sc
plt.figure(figsize=(20,7))
plt.title("Dendograms")

sc.dendrogram(sc.linkage(pca_saled,method='ward'))
plt.title('Dendogram')
plt.xlabel('sample Index')
plt.ylabel('Eucledian distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering 
cluster = AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage='ward')
cluster_labels = cluster.fit(pca_saled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_saled[:,0],pca_saled[:,1],c =cluster.labels_)
```
